```python
url='https://finance.naver.com/marketindex/exchangeList.naver'
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
res=requests.get(url).text
```

```python
print(res)
```

```python
soup=BeautifulSoup(res,'lxml')
result=soup.select('td.tit > a')
```

```python
print(result)
```

```python
name=[]
for item in result:
    name.append(item.string.strip())
```

```python
print(name)
```

```python
result2=soup.select("td[class='sale']")
```

```python
print(result2)
```

```python
salestandard=[]
for item in result2:
    salestandard.append(float(item.string.replace(',','')))
print(list(zip(name,salestandard)))
```

```python
import requests
from bs4 import BeautifulSoup
url2='http://www.cgv.co.kr/movies/'
res3=requests.get(url2).text
soup=BeautifulSoup(res3,'lxml')
result3=soup.select("strong.title")
moviename=[]
for item in result3:
    moviename.append(item.string.strip())
print(moviename)
result4=soup.select('span.thumb-image img')
img=[]
for item2 in result4:
    img.append(item2.string)
print(img)
```
