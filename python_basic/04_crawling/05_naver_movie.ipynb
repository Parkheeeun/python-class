```python
import requests
from bs4 import BeautifulSoup
import re

url='https://movie.naver.com/movie/point/af/list.naver?&page={}'
reviewlist=[]

for i in range(1,11):
    print(url.format(i))
    res=requests.get(url.format(i)).text
    soup=BeautifulSoup(res,'lxml')
    table=soup.find('table',class_='list_netizen')
    for i,r in enumerate(table.select('tbody tr')):
        # print(i)
        # print(r)
        for j,c in enumerate(r.find_all('td')):
            if j==0:
                recode1=int(c.text.strip())
                print('글번호:',recode1)
            elif j==1:
                recode2=c.select_one('a').text.strip()
                print('영화명:',recode2)
                recode3=c.select_one('em').text.strip()
                print('영화평점:',recode3)
                recode4=c.text
                recode4=recode4.replace(recode2,'')
                recode4=recode4.replace('신고','')
                recode4=re.sub('별점 - 총 10점 중[0-9]{1,2}','',recode4).strip()
                print('감상평:',recode4)
                try:
                    movie_dic={
                        '영화명':recode2,
                        '영화평점':recode3,
                        '감상평':recode4
                    }
                    reviewlist.append(movie_dic)
                except:
                    pass
print(reviewlist)
            
```
